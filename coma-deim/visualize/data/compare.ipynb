```python

attr_auc = []
attr_nll = []
edge_auc = []
edge_nll = []
persona = 12



attr_auc_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_attr_auc.npy".format(persona)
attr_nll_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_attr_nll.npy".format(persona)
edge_auc_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_edge_auc.npy".format(persona)
edge_nll_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_edge_nll.npy".format(persona)

attr_auc.append(np.load(attr_auc_path))
attr_nll.append(np.load(attr_nll_path))
edge_auc.append(np.load(edge_auc_path))
edge_nll.append(np.load(edge_nll_path))

attr_auc_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/k-mean/persona={}/proposed_attr_auc.npy".format(persona)
attr_nll_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/k-mean/persona={}/proposed_attr_nll.npy".format(persona)
edge_auc_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/k-mean/persona={}/proposed_edge_auc.npy".format(persona)
edge_nll_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/k-mean/persona={}/proposed_edge_nll.npy".format(persona)

attr_auc.append(np.load(attr_auc_path))
attr_nll.append(np.load(attr_nll_path))
edge_auc.append(np.load(edge_auc_path))
edge_nll.append(np.load(edge_nll_path))

attr_mean = np.mean(attr_auc,axis=1)
attr_max = np.max(attr_auc,axis=1)
attr_min = np.min(attr_auc,axis=1)
edge_mean = np.mean(edge_auc,axis=1)
edge_max = np.max(edge_auc,axis=1)
edge_min = np.min(edge_auc,axis=1)
```

```python
fig, ax = plt.subplots(1)
t = [1,2,3,4,5]

ax.plot(t, edge_mean[0], lw=2, label='proposed', color='blue')
ax.plot(t, edge_mean[1], lw=2, label='k-mean', color='red')





ax.fill_between(t, edge_min[0], edge_max[0], facecolor='blue', alpha=0.1)
ax.fill_between(t, edge_min[1], edge_max[1], facecolor='red', alpha=0.1)


ax.set_title(r'edge auc persona={}'.format(persona))
ax.legend(loc='upper left')
ax.set_xlabel('time')
ax.set_ylabel('accuracy')
ax.set_xticks([1,2,3,4,5])
```
