```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cbook as cbook
```

```python
def plot_data(data_name,plot_type):   
    fig, ax = plt.subplots(1)
    colors = ["#FF4B00","#03AF7A","#F6AA00","#FFF100","#000000","#4DC4FF"]

    t=0

    if data_name == "NIPS":
        persona_list = [3,5,8,12,16]
    else:
        persona_list = [5,25,50]

    for p in persona_list:
        attr_auc = []
        attr_nll = []
        edge_auc = []
        edge_nll = []
       


        path = "../experiment_data/complete/{}/persona={}".format(data_name,p)

        attr_auc_path = path + "/proposed_attr_auc.npy"
        attr_nll_path = path + "/proposed_attr_nll.npy"
        edge_auc_path = path + "/proposed_edge_auc.npy"
        edge_nll_path = path + "/proposed_edge_nll.npy"

        attr_auc.append(np.load(attr_auc_path))
        attr_nll.append(np.load(attr_nll_path))
        edge_auc.append(np.load(edge_auc_path))
        edge_nll.append(np.load(edge_nll_path))

        attr_mean = np.mean(attr_auc,axis=1)
        attr_max = np.max(attr_auc,axis=1)
        attr_min = np.min(attr_auc,axis=1)
        edge_mean = np.mean(edge_auc,axis=1)
        edge_max = np.max(edge_auc,axis=1)
        edge_min = np.min(edge_auc,axis=1)

        time = [1,2,3,4,5]

        if plot_type == "edge":
            ax.plot(time, edge_max[0], lw=2, label="Proposed:"+str(p), color=colors[t])
            ax.fill_between(time, edge_min[0], edge_max[0], facecolor=colors[t], alpha=0.1)
        else:
            ax.plot(time, attr_max[0], lw=2, label="Proposed;"+str(p), color=colors[t])
            ax.fill_between(time, attr_min[0], attr_max[0], facecolor=colors[t], alpha=0.1)
        t += 1

    time = [1,2,3,4,5]

    #ax.fill_between(time, edge_min[0], edge_max[0], facecolor="blue", alpha=0.1)

    ax.set_xticks([1,2,3,4,5])
    ax.set_title(r'{} auc [{}]'.format(plot_type,data_name))
    ax.legend(loc='upper left',bbox_to_anchor=(1, 1))
    ax.set_xlabel('time')
    ax.set_ylabel('accuracy')
```

```python
def plot_complete(ax,data_name,plot_type):
    #both

    attr_auc = []
    attr_nll = []
    edge_auc = []
    edge_nll = []

   

    path = "../experiment_data/complete/{}".format(data_name)

    attr_auc_path = path + "/proposed_attr_auc.npy"
    attr_nll_path = path + "/proposed_attr_nll.npy"
    edge_auc_path = path + "/proposed_edge_auc.npy"
    edge_nll_path = path + "/proposed_edge_nll.npy"

        

    attr_auc.append(np.load(attr_auc_path))
    attr_nll.append(np.load(attr_nll_path))
    edge_auc.append(np.load(edge_auc_path))
    edge_nll.append(np.load(edge_nll_path))

    attr_mean = np.mean(attr_auc,axis=1)
    attr_max = np.max(attr_auc,axis=1)
    attr_min = np.min(attr_auc,axis=1)
    edge_mean = np.mean(edge_auc,axis=1)
    edge_max = np.max(edge_auc,axis=1)
    edge_min = np.min(edge_auc,axis=1)

    time = [1,2,3,4,5]

    if plot_type == "edge":
        ax.plot(time, edge_mean[0], lw=2, label="Complete", color="#005AFF")
        ax.fill_between(time, edge_min[0], edge_max[0], facecolor="#005AFF", alpha=0.1)

    else:
        ax.plot(time, attr_mean[0], lw=2, label="Complete", color="#005AFF")
        ax.fill_between(time, attr_min[0], attr_max[0], facecolor="#005AFF", alpha=0.1)
```

```python
def plot_persona(ax,data_name,del_type,plot_type,percent):   
    
    colors =["#FF4B00","#03AF7A","#F6AA00","#FFF100","#000000","#4DC4FF"]



    if data_name == "NIPS":
        persona_list = [3,5,8,12,16]
    else:
        persona_list = [5,25,50]
    t=0
    if data_name == "NIPS":
        repeat = 10
    else:
        repeat =1
        
    for persona in persona_list:
       
        attr_auc = []
        attr_nll = []
        edge_auc = []
        edge_nll = []



        for attempt in range(repeat):

            path = "../experiment_data/imcomplete/{}/t=4/{}/percent={}/attempt={}/persona={}".format(data_name,del_type,percent,attempt,persona)

            attr_auc_path = path + "/proposed_attr_auc.npy"
            attr_nll_path = path + "/proposed_attr_nll.npy"
            edge_auc_path = path + "/proposed_edge_auc.npy"
            edge_nll_path = path + "/proposed_edge_nll.npy"

              

            attr_auc.append(np.load(attr_auc_path))
            attr_nll.append(np.load(attr_nll_path))
            edge_auc.append(np.load(edge_auc_path))
            edge_nll.append(np.load(edge_nll_path))

            attr_mean = np.mean(attr_auc,axis=1)
            attr_max = np.max(attr_auc,axis=1)
            attr_min = np.min(attr_auc,axis=1)
            edge_mean = np.mean(edge_auc,axis=1)
            edge_max = np.max(edge_auc,axis=1)
            edge_min = np.min(edge_auc,axis=1)

        time = [1,2,3,4,5]

        if plot_type == "edge":
            ax.plot(time, edge_mean[0], lw=2, label="Persona:"+str(persona), color=colors[t])
            ax.fill_between(time, edge_min[0], edge_max[0], facecolor=colors[t], alpha=0.1)

        else:
            ax.plot(time, attr_mean[0], lw=2, label="Persona:"+str(persona), color=colors[t])
            ax.fill_between(time, attr_min[0], attr_max[0], facecolor=colors[t], alpha=0.1)
        
        t += 1



    #plot_complete(ax,data_name,plot_type)

    ax.set_xticks([1,2,3,4,5])
    ax.set_title(r'{} auc Delete:{} [{} Percent:{}]'.format(plot_type,del_type,data_name,percent))
    ax.legend(loc='upper left',bbox_to_anchor=(1, 1))
    ax.set_xlabel('time')
    ax.set_ylabel('accuracy')
```

```python
def change_persona(data_name,del_type,plot_type):

    for p in [5,15,30,50,75]:

        fig, ax = plt.subplots(1)
        plot_persona(ax,data_name,del_type,plot_type,p)
```
