```python
attr_auc = []
attr_nll = []
edge_auc = []
edge_nll = []




attr_auc_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/64_5/persona = 32/proposed_attr_auc.npy"
attr_nll_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/64_5/persona = 32/proposed_attr_nll.npy"
edge_auc_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/64_5/persona = 32/proposed_edge_auc.npy"
edge_nll_path = "//Users/matsumoto-hirotomo/coma/experiment_data/NIPS/64_5/persona = 32/proposed_edge_nll.npy"

attr_auc.append(np.load(attr_auc_path))
attr_nll.append(np.load(attr_nll_path))
edge_auc.append(np.load(edge_auc_path))
edge_nll.append(np.load(edge_nll_path))

for i in [3,4,5,6,8,12]:
    attr_auc_path = "/Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_attr_auc.npy".format(i)
    attr_nll_path = "/Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_attr_nll.npy".format(i)
    edge_auc_path = "/Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_edge_auc.npy".format(i)
    edge_nll_path = "/Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_edge_nll.npy".format(i)

    attr_auc.append(np.load(attr_auc_path))
    attr_nll.append(np.load(attr_nll_path))
    edge_auc.append(np.load(edge_auc_path))
    edge_nll.append(np.load(edge_nll_path))


attr_mean = np.mean(attr_auc,axis=1)
attr_max = np.max(attr_auc,axis=1)
attr_min = np.min(attr_auc,axis=1)
edge_mean = np.mean(edge_auc,axis=1)
edge_max = np.max(edge_auc,axis=1)
edge_min = np.min(edge_auc,axis=1)
```
