```python

import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cbook as cbook
```

```python


fig, axes = plt.subplots(nrows=32, ncols=2, figsize=(10, 100))
for i in range(32):

    attr_auc = []
    attr_nll = []
    edge_auc = []
    edge_nll = []
    

    persona=5
    attr_auc_path = "/Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_attr_auc.npy".format(persona)
    attr_nll_path = "/Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_attr_nll.npy".format(persona)
    edge_auc_path = "/Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_edge_auc.npy".format(persona)
    edge_nll_path = "/Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/persona={}/proposed_edge_nll.npy".format(persona)

    attr_auc.append(np.load(attr_auc_path))
    attr_nll.append(np.load(attr_nll_path))
    edge_auc.append(np.load(edge_auc_path))
    edge_nll.append(np.load(edge_nll_path))

   

    path = "/Users/matsumoto-hirotomo/coma/experiment_data/NIPS/200_20/incomplete/"
    attr_auc_path = path+"drop={}/persona=5/proposed_attr_auc.npy".format(i)
    attr_nll_path = path+"drop={}/persona=5/proposed_attr_nll.npy".format(i)
    edge_auc_path = path+"drop={}/persona=5/proposed_edge_auc.npy".format(i)
    edge_nll_path = path+"drop={}/persona=5/proposed_edge_nll.npy".format(i)

    attr_auc.append(np.load(attr_auc_path))
    attr_nll.append(np.load(attr_nll_path))
    edge_auc.append(np.load(edge_auc_path))
    edge_nll.append(np.load(edge_nll_path))

    attr_mean = np.mean(attr_auc,axis=1)
    attr_max = np.max(attr_auc,axis=1)
    attr_min = np.min(attr_auc,axis=1)
    edge_mean = np.mean(edge_auc,axis=1)
    edge_max = np.max(edge_auc,axis=1)
    edge_min = np.min(edge_auc,axis=1)
    print(edge_max)


    t = [1,2,3,4,5]

    axes[i,0].plot(t, edge_mean[0], lw=2, label='persona5', color='blue')
    axes[i,0].plot(t, edge_mean[1], lw=2, label='Incomplete', color='red')





    axes[i,0].fill_between(t, edge_min[0], edge_max[0], facecolor='blue', alpha=0.1)
    axes[i,0].fill_between(t, edge_min[1], edge_max[1], facecolor='red', alpha=0.1)


    axes[i,0].set_title(r'edge auc {}'.format(persona))
    axes[i,0].legend(loc='upper left')
    axes[i,0].set_xlabel('time')
    axes[i,0].set_ylabel('accuracy')




    axes[i,1].plot(t, attr_mean[0], lw=2, label='proposed', color='blue')
    axes[i,1].plot(t, attr_mean[1], lw=2, label='k-mean', color='red')
    axes[i,1].set_xticks([1,2,3,4,5])


    axes[i,1].fill_between(t, attr_min[0], attr_max[0], facecolor='blue', alpha=0.1)
    axes[i,1].fill_between(t, attr_min[1], attr_max[1], facecolor='red', alpha=0.1)

    axes[i,1].set_title(r'attr auc persona={}'.format(persona))
    axes[i,1].legend(loc='upper left')
    axes[i,1].set_xlabel('time')
    axes[i,1].set_ylabel('accuracy')

plt.tight_layout()
plt.show()
```
